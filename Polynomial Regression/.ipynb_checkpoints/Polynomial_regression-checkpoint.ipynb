```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('Position_Salaries.csv')
x = df.iloc[:,1:-1].values
y = df.iloc[:,-1].values
```

```python
linear_reg = LinearRegression()
linear_reg.fit(x,y)
linear_y_pred = lin_reg.predict(x)
```

```python
polynomial_reg = PolynomialFeatures(degree=4)
x_poly =polynomial_reg.fit_transform(x)
lin2_reg = LinearRegression()
lin2_reg.fit(x_poly,y)
poly_y_pred = lin2_reg.predict(x_poly)
```

```python
plt.scatter(x,y, color ='green')
plt.plot(x,linear_y_pred,color ='yellow')
plt.title('Linear Regression')
plt.xlabel('position')
plt.ylabel('salary')
plt.show
```

```python
plt.scatter(x,y,color='green')
plt.plot(x,poly_y_pred,color='blue')
plt.title('Polynomial Regression')
plt.xlabel('position')
plt.ylabel('salary')
```

```python
r2_linear = r2_score(y,linear_y_pred)
print(r2_linear)
```

```python
r2_polynomial = r2_score(y,poly_y_pred)
print(r2_polynomial)
```
